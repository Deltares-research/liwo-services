```python
import requests
import json
```

```python
# base = "https://basisinformatie-overstromingen.nl/"
# base = "http://34.117.7.205/"
# base = "http://localhost:5001/"
base = "http://127.0.0.1:5000/"
```

```python
def get_dict_d(res):
    data = res.content.decode("utf-8")
    return json.loads(json.loads(data.strip())["d"])


def get_dict(res):
    data = res.content.decode("utf-8")
    return json.loads(data.strip())
```

```python
url = base + "api/liwo.ws/Maps.asmx/GetLayerSet"
res = requests.post(url, json={"id": 6})
get_layer_set = get_dict_d(res)
get_layer_set
```
